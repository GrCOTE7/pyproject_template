$$
h(n) =
\begin{cases}
n & \text{si } n \leq 2 \\
\sum_{i=0}^{n-1} (n+i) \cdot h(i) & \text{sinon}
\end{cases}
$$


```python
def h(n):
    """
    Compute the recursive function h(n).

    The function is defined as:

    h(n) = n                  si n <= 2
    h(n) = somme_{i=0}^{n-1} (n+i) * h(i)   sinon

    Parameters
    ----------
    n : int
        Natural number (n ≥ 0).

    Returns
    -------
    int
        Value of the recursive function h(n).

    Notes
    -----
    - Base case: returns `n` directly if `n <= 2`.
    - Recursive case: computes the sum of `(n + i) * h(i)` for all `i` in `[0, n-1]`.

    Examples
    --------
    >>> h(0) => 0
    >>> h(1) => 1
    >>> h(2) => 2
    >>> h(3) => 14
    >>> h(4) => 115
    """

    if n <= 2:
        return [n,n]
    return [
        sum((n + i) * h(i)[0] for i in range(n)),               # Calcul h() normal
        (2 * n - 1) * h(n - 1)[1] + (2 * n - 2) * h(n - 2)[1],  # Calcul Cyprien
        # https://discord.com/channels/1056923339546968127/1453744085641855158/1453744085641855158
    ]

if __name__== "__main__":
    print("  n =>      h(n)    | h(n) de Cyprien", sep="\n")
    print ('-' * 39)
    print(*[f"{i:> 3} => {h(i)[0]:> 12} | {h(i)[1]:> 12}" for i in range(11)], sep="\n")
```

```python
import matplotlib.pyplot as plt

# Calcul des valeurs pour n de 0 à 10
N = 10
x_vals = list(range(N+1))
y_vals = [h(n) for n in x_vals]

# Tracé du graphique 
plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, marker='o', linestyle='-', color='royalblue', label='h(n)')
plt.title("Représentation logarithmique de la fonction récursive h(n)")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.yscale('log')
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math


# Définition récursive de h(n)
def h(n):
    if n <= 2:
        return n
    return sum((n + i) * h(i) for i in range(n))


# Calcul des valeurs
N = 10
x_vals = list(range(N + 1))
y_h = [h(n) for n in x_vals]
y_fact = [math.factorial(n) for n in x_vals]
y_pow2 = [2**n for n in x_vals]

# Tracé
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_h, marker="o", linestyle="-", color="royalblue", label="h(n)")
plt.plot(x_vals, y_fact, marker="s", linestyle="--", color="darkorange", label="n!")
plt.plot(x_vals, y_pow2, marker="^", linestyle="-.", color="forestgreen", label="2^n")

plt.title("Comparaison de h(n), n! et 2^n")
plt.xlabel("n")
plt.ylabel("Valeurs")
plt.yscale("log")  # échelle logarithmique
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
